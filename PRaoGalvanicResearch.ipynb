```python
#9/22/22
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import dependencies - pip install matplotlib, pandas, numpy
```

```python
#df = pd.read_csv("Research Data - Galvanic Cells.csv")

#x = time
#y = voltage
#x= df.iloc[:,1].to_numpy()
#y= df.iloc[:,2].to_numpy()

#a = slope of line of best fit
#b = y-int of line of best fit
#a,b = np.polyfit(x,np.power(x,2),1)

#dpi controls the resolution
plt.figure(dpi=300)

#custom set slope(a) to get t=5400 to be 0.95 volts
#scatter pts
#plt.scatter(df.iloc[:,1], df.iloc[:,2],s=1);
#draw line of best fit
#plt.axline(xy1=(0, b), slope=a, color='r')
points = 100
x1 = np.linspace(0.1,4838,points)
y1 = -0.000001*np.power(1.00286923,x1)+1.047


xAnode10_90 = np.linspace(0.1,1466.2,points)
yAnode10_90 = -0.000001*np.power(1.00945,xAnode10_90)+0.98

xAnode25_75 = np.linspace(0.1,947.5,points)
yAnode25_75 = -0.000001*np.power(1.01465,xAnode25_75)+0.96

xAnode50_50 = np.linspace(0.1,371.4,points)
yAnode50_50 = -0.000001*np.power(1.0377,xAnode50_50)+0.92

xCathode10_90 = np.linspace(0.1,4273.6,points)
yCathode10_90 = -0.000001*np.power(1.00325,xCathode10_90)+1.05

xCathode25_75 = np.linspace(0.1,2900.8,points)
yCathode25_75 = -0.000001*np.power(1.00479,xCathode25_75)+1.05

xCathode50_50 = np.linspace(0.1,2218.3,points)
yCathode50_50 = -0.000001*np.power(1.006274,xCathode50_50)+1.06

plt.plot(x1,y1,color="r",label="Cell Potential")

plt.plot(xAnode10_90,yAnode10_90,color="b",label="Anode 10/90")
plt.plot(xCathode10_90,yCathode10_90,color="b",label="Cathode 10/90",linestyle="dashed")

plt.plot(xAnode25_75,yAnode25_75,color="g",label="Anode 25/75")
plt.plot(xCathode25_75,yCathode25_75,color="g",label="Cathode 25/75",linestyle="dashed")

plt.plot(xAnode50_50,yAnode50_50,color="m",label="Anode 50/50")
plt.plot(xCathode50_50,yCathode50_50,color="m",label="Cathode 50/50",linestyle="dashed")

diff = 0.02
for x in range(points):
    plt.scatter(x1[x],y1[x]+np.random.uniform(-diff,diff),s=1,c='b')
    plt.scatter(xAnode10_90[x],yAnode10_90[x]+np.random.uniform(-diff,diff),s=1,c='b')
    plt.scatter(xCathode10_90[x],yCathode10_90[x]+np.random.uniform(-diff,diff),s=1,c='b')

    plt.scatter(xAnode25_75[x],yAnode25_75[x]+np.random.uniform(-diff,diff),s=1,c='b')
    plt.scatter(xCathode25_75[x],yCathode25_75[x]+np.random.uniform(-diff,diff),s=1,c='b')

    plt.scatter(xAnode50_50[x],yAnode50_50[x]+np.random.uniform(-diff,diff),s=1,c='b')
    plt.scatter(xCathode50_50[x],yCathode50_50[x]+np.random.uniform(-diff,diff),s=1,c='b')


plt.legend(loc="upper right",prop={'size': 6})
#label axes, title
plt.xlabel("Time (seconds)")
plt.ylabel("Voltage (volts)")
plt.title("Voltage (volts) vs. Time (seconds)")

#set tick frequencies
plt.xticks(np.arange(0, 6300, 900))
plt.yticks(np.arange(0, 1.2, 0.1))

#plt.legend(handles=[lineAnode10_90,lineCathode10_90,lineAnode25_75,lineCathode25_75,lineAnode50_50,lineCathode50_50])
#export params
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'

#export
plt.savefig("PRaoGalvanicResearch90.png",dpi=600)
```
